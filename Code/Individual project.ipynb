```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df1 = pd.read_csv('Mental Health Social Media Dataset.csv')
df1
```

```python
df1 = df1.drop(columns = ['Serial_Number', 'Timestamp', 'What_is_your_age', 'Gender', 'Relationship_Status' , 'Occupation_Status', 'What_type_of_organizations_are_you_affiliated_with','What_social_media_platforms_do_you_commonly_use', 'How_do_you_feel_about_these_comparisons_generally_speaking','How_often_do_you_compare_yourself_to_other_successful_people', 'Do_you_find_it_difficult_to_concentrate_on_things', 'Do_you_feel_restless_if_you_havent_used_Social_media', 'Do_you_find_yourself_using_Social_media_without_reason', 'Do_you_get_distracted_by_Social_media_when_you_are_busy'])
df1               
```

```python
 df1[' What_is_the_average_time_you_spend_on_social_media_every_day'] = df1[' What_is_the_average_time_you_spend_on_social_media_every_day'].replace({
        'Between 1 and 2 hours': 1.5,
        'Between 2 and 3 hours': 2.5,
        'Between 3 and 4 hours': 3.5,
        'Between 4 and 5 hours': 4.5,
        'Less than an Hour': 0.5,
        'More than 5 hours': 6
    })
df1
```

```python
df1 = df1.drop([134, 9, 54,]) #Removing the ones who responded 'no' to 'Do_you_use_social_media' 
df1
```

```python
mode_media = df1[' What_is_the_average_time_you_spend_on_social_media_every_day'].mode()
mode_media
```

```python
average_distracted = df1['How_easily_distracted_are_you'].mean()
average_distracted
```

```python
std_distracted = df1['How_easily_distracted_are_you'].std()
std_distracted
```

```python
average_worried = df1['How_much_are_you_bothered_by_worries'].mean()
average_worried
```

```python
std_worried = df1['How_much_are_you_bothered_by_worries'].std()
std_worried
```

```python
average_validation = df1['Do_you_look_to_seek_validation_from_features_of_social_media'].mean()
average_validation
```

```python
std_validation = df1['Do_you_look_to_seek_validation_from_features_of_social_media'].std()
std_validation
```

```python
average_depressed = df1['How_often_do_you_feel_depressed_or_down'].mean()
average_depressed
```

```python
std_depressed =df1 ['How_often_do_you_feel_depressed_or_down'].std()
std_depressed
```

```python
plt.hexbin(df1[' What_is_the_average_time_you_spend_on_social_media_every_day'], 
           df1['How_often_do_you_feel_depressed_or_down'], gridsize=30, cmap='Blues')
plt.colorbar()
plt.xlabel('Average Time on Social Media (hours)')
plt.ylabel('Depression Frequency')
plt.title('Hexbin Plot: Social Media Time vs Depression Frequency')
plt.show()
```

```python
plt.hexbin(df1[' What_is_the_average_time_you_spend_on_social_media_every_day'], 
           df1['How_much_are_you_bothered_by_worries'], gridsize=30, cmap='Greens')
plt.colorbar()
plt.xlabel('Average Time on Social Media (hours)')
plt.ylabel('Worry Frequency')
plt.title('Hexbin Plot: Social Media Time vs Worry Frequency')
plt.show()
```

```python
plt.hexbin(df1[' What_is_the_average_time_you_spend_on_social_media_every_day'], 
           df1['How_easily_distracted_are_you'], gridsize=30, cmap='Purples')
plt.colorbar()
plt.xlabel('Average Time on Social Media (hours)')
plt.ylabel('Distraction Frequency')
plt.title('Hexbin Plot: Social Media Time vs Distraction Frequency')
plt.show()
```
